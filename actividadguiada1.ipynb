# Torres de hanoi
## Algoritos recursivos
Un algoritmo recursivo es un mtodo de programacin en que la funcin se invoca a si isma para resolver una versin reducida del mismo problema.
Este proceso contina hasta que el problema se simplifica  lo suficinete como para resolverse directamente, sin necesidad de invocaciones adicionales.
La idea clave de la recursin es que el problema se divide en partes ms pequeas y manejables, lo que hace ms facil resolverlo.

### Ejercicio Torres de Hanoi
Se llama las **Torres de hanoi**. Te dan un conjunto de tres varillas y n discos, con cada disco de un tamao diferente. LLamaremos a las varillas A, B, y numeramos los discos desde 1, siendo el disco ms pequeo, hasta n, siendo el disco ms grande. Al principio, todos los n discos estn en la varilla A, en orden de tamao decreciente de la parte inferior a la parte superior, de modo que el disco n est en la parte inferior  y el disco 1 est en la parte superior. Aqu esta como se ven las torres de Hanoi para n=4 discos:



El objetivo del ejercicio es pasar todos los n discos de la varilla A a la varilla B.


**Debemos seguir las siguientes reglas:**
* Podemos mover solamanete un disco a la vez
* Ningn disco puede estar encima de un disco ms pequeo. Por ejemplo, si el disco 3 est en una varilla, entonces todos los discos debajo del disco 3 deben tener nmeros mayores que 3.

## Pasos a seguir

Siempre puedes mover el disco1 de la varilla A a la varilla B, porqu sabes que cualquier disco debajo debe ser mayor.  Y no hay nada especial acerca de las varillas A y B. Puedes mover el disco de la varilla B a la varilla C si lo deseas, o de la varilla C a la varilla A, o de cualquier varilla. Resolver el problema de las Torres de Hanoi con un deisco es trivial, y requiere mover el nico disco solamente una vez.

<br>La sucesin de Fibonacci es conocida desde hace miles de aos, pero fue Fibonacci (Leonardo de Pisa) quien le dio a conocer al utilizarla para resolver un problema.<br>

## Trmino general
La sucesin de Fibonacci es una sucesin definida por recurrencia. Esto significa que para calcular un trmino de sucesin se necesitan los trminos que le preceden.<br>
Se proporcionan los dos primeros trminos: $a_0 = 0$ y $a_1 = 1$. Los siguinetes se calculan con la siguinete frmula:<br>

$$
a_n+1 = a_n-1 + a_n, n>= 1
$$
<br>
**Nota:** el primer trmino que proporciona la frmula es **$a_2$** (porque **$n$** tiene que ser mayor o igual que **1**).<br>
Por esta razn, se definen **$a_0$** y **$a_1$** con anterioridad.



```python
#Calculo del trmino n-simo de la sucesin de Fibonacci
def Fibonacci(N:int):
    if N<2:
        return 1
    else:
        return Fibonacci(N-1)+Fibonacci(N-2)
        
```

```python
# Funcin para generar la serie de Fibonacci hasta el trmino n
def serie_fibonacci(N):
    serie = []
    for i in range (N + 1):
        serie.append(Fibonacci(i))
    return serie
```

# Frmula de Binet
**La frmula de Binet** es una frula explicita y cerrada que se utiliza para hallar rl $n$ trmino n-simo de la sucesin de Fibonacci. Recibe este nombre porque fue derivada por el matemtico Jacques Philippe Marie Binet, aunque ya era conocidapor Abraham de Moivre.

**Frmula**
Si $F_n$ es el $n^0$ nmero de Fibonacci, entonces:<br>
$$
F_n = \frac {1}{\sqrt{a}}((\frac{1+\sqrt{5}}{2})^n -(\frac{1-\sqrt{5}}{2})^n)
$$

```python
#importamos la libreria math
import math
def formula_binet(n):
    #calculamos el nmero ureo
    phi = (1 + math.sqrt(5)) / 2
    psi = (1 - math.sqrt(5)) / 2

    resultado = (phi**n - psi**n) / math.sqrt(5)
    return round(resultado)
```

```python
#implementamos una funcin que imprime los primeros n niveles en forma de piramide
def piramide(numero_niveles):
    #Calculamos lso trminos de Fibonnaci necesario
    total_numeros = numero_niveles * (numero_niveles +1)//2
    fibos=[formula_binet(i) for i in range(total_numeros)]
    index = 0
    for fila in range(1, numero_niveles+1):
        #espacios
        print (" " * (numero_niveles-fila) * 3, end="")
        for _ in range(fila):
            print(f"{fibos[index]:<4}", end=" ")
            index+=1
        print()
```

# Devolucin de cambio por tcnica voraz

Un algoritmo voraz(greddy) es un algoritmo que encuentra una solucin globalmente ptima a un problema a basde de hacer elecciones localmente ptimas. Es decir: el algoritmo siempre hace lo que "parece" mejor en cada momento, sin tener nunca que reconsiderar sus decisiones, y acaba llegando directamente a la mejor solucin posible.

```python
def cambio_moneda(N, SM):
    SOLUCION = [0] * len(SM)
    ValorAcumulado= 0

    for i,valor in enumerate(SM):
        monedas = (N-ValorAcumulado)//valor
        SOLUCION[i]= monedas
        ValorAcumulado = ValorAcumulado + monedas*valor

        if ValorAcumulado == N:
            return SOLUCION
    
```

Para complementar el concepto de algoritmos voraces eh conseguido un ejercicio.<br>
Hay $M(M<=10^5)$ farolas en la posisicin $y_1,...,y_M$ de una recta y $N(N<=10^5)$ puntos $x_1,...,x_N$. Cada farola tiene un radio de iluminacin $r_i$, tal que la i-sima farola ilumina puntos en el intervalo $[y_i-r_i,y_i+r_i]$. Se requiere encender el mnimo nmero de farolas tales que cada uno de los $N$ puntos $x_1,...x_N$ est iluminado por al menos una farola. Encuentra este mnimo nmero. <br>

### Solucin
* Se convierte cada farola en un intervalo de cobertura:
  intervalo= $[y_i -r_i, y_i+r_i]$
* Ordenamos estos intervalos por el extremo izquierdo o por el final
* Ordenamos los puntos x.
* Para cada punto x:
  * Mientras haya farolas cuyo intervalo **empieza antes o justo en x**, guardamos la que ms lejos llegue
  * Encendemos esa farola, y saltamos al siguiente punto que no est cubierto poe ella.



```python
#Importamos la biblioteca de matplotlib
import matplotlib.pyplot as plt
#Implementamos una funcin 
def min_farolas(puntos, farolas):
    #Convertimos las farolas en intervalos [inicio y fin]
    intervalos = []
    for idx, (y, r) in enumerate(farolas):
        #aadimos los intervalos a las farolas
        intervalos.append((y-r,y+r,idx))
    # Ordenamos los intervalos por el inicio
    intervalos.sort()

    #Ordenamos los puntos que se deben cubrir
    puntos.sort()

    #Declaramos e inicializamos el nuemro de farolas encendidas
    resultado=0
    encendidas=[]
    #Inicializamos el indice de lso intervalos
    i=0
    #Sacamos la longitud de los intervalos y lo guardamos en una variable n 
    n=len(intervalos)
    #Inicializamos los indices para los puntos
    j=0
    #Creamos un cliclo repetitivo 
    while j < len(puntos):
        x=puntos[j]
        max_cobertura = -1
        mejor_farola=-1
        
        #Buscamos la mejor farola que cubrea c
        while i < n and intervalos[i][0] <=x:
            if intervalos[i][1] > max_cobertura:
                max_cobertura= max(max_cobertura, intervalos[i][1])
                # Guardamos el indice
                mejor_farola=intervalos[i][2]
            i+=1
        if max_cobertura <x:
            #No hay ninguna farola que cubra el punto x
            return -1, [] # No se puede cubrir todos los puntos
        #Encedemos una farola que cubre hasta max-cobertura
        resultado += 1
        encendidas.append(farolas[mejor_farola])
        #saltamos todos lo puntos ya cubiertos
        while j< len(puntos) and puntos[j] <= max_cobertura:
            j+=1
        
    return resultado, encendidas
    
    
```

```python
# 3 farolas  con sus posiciones y radios
farolas = [(2,2), (6,1), (10,2)]

# puntos que deben ser cubiertos
puntos = [1,3,5,6,11]

minimo, usadas = min_farolas(puntos, farolas)
print("Minimo de farolas", minimo)
print("Farolas utilizadas: ")
for pos, radio in usadas:
    print(f" Posicin: {pos}, Radio: {radio}")
```

```python
# Visualizacin grfica 
fig, ax = plt.subplots(figsize=(10,2))
#Dibujamos los puntos rojo
for p in puntos:
    # Punto a cubrir
    #Colocamos los puntos ligeramente arriba de la linea
    ax.plot(p, 0.02, 'ro', zorder=3)
    ax.text(p, 0.05, f"{p}", ha='center', fontsize=9, zorder=4)

#dibujar farolas como intervalos
for pos, radio in usadas:
    inicio = pos - radio
    fin = pos + radio
    #barra de cobertura
    ax.plot([inicio,fin],[0,0], lw=8, alpha=0.4, zorder=1)
    #centro de la farola
    ax.plot(pos, 0, 'bo', zorder=2)
    ax.text(pos, -0.02, f"({pos},{radio})", ha='center', fontsize=8, color='blue', zorder=2)

#Configuramos el grfico
ax.set_yticks([])
ax.set_xlabel("Posicin en la recta")
ax.set_title("Cobertura de farolas sobre puntos ")
ax.grid(True)
plt.subplots_adjust(top=0.85, bottom=0.25)
plt.show()
```

## Por qu es una solucin voraz?
Porque en cada se elige la mejor opcin local posible (la farola que ms cubre el siguiente punto no cubierto), sin irar al futuro.<br>
En este tipo de problemas, la eleccin local resulta ser globalmente ptima, siempre y cuando lso intervalos estn ordenados.

# N-Reinas por tcnica vuelta atrs
### Qu es la Tcnica de Vuelta Atrs?<br>
La vuelta atrs (backtracking) es una estrategia de resolucin de problemas basada en la bsqueda sistemtica de soluciones en un espacio de decisiones. Se utiliza cuando el problema requiere construir una solucin paso a paso, y en cada paso se deben validar ciertas condiciones para determinar si se puede continuar o si es necesario retroceder(backtrack) y probar otras alternativas.<br>
Es una tcnica especialmente til en problemas que involucran:
* Combinatorias (permutaciones, combinaciones)
* Configuraciones (como tableros, secuencias o asignaciones)
* Satisfaccin de restricciones (como en juegos, puzzles o tareas de asigancin)

## Principios de Funcionamiento
EL algoritmo de vuelta atrs explora todas las posibles soluciones de manera estrucutrada utilizando recursin. En cada paso.<br>
* Se genera una decisin parcil.
* Se verifica si es vlida bajo las restricciones del problema.
* Se la solucin parcial es vlida, se contina con la siguiente decisin.
* Si se detecta que no puede conducir a una solucin completa vlida, se deshace ( se retorcede) y se prueba otra opcin
* Este proceso se asemeja a un rbol de decisin donde el algoritmo recorre ramas y vueleve al nodo anterior cuando encuentra un camino invalido.

# Caracteristicas del Algoritmo
* **Recursivo:** la solucin se construyr paso a paso
* **Eficiente** frente a fuerza bruta, ya que descarta cambios no viables temprano
* **Explcitamente controlado:**

# Descripcin del Problema
El problema de las N-Reinas consiste en ubicar N reinas en tablero de agedrez d NxN, d emodo que ninguna reina ataque a otra. En trminos prcticos, se deben cumplir las siguientes condiciones:

* Ninguna reina debe compartir la misma fila.
* Ninguna reina debe compartir la misma columna.
* Ninguna reina debe ubicarse en la mima dagonal qeu otra.
Este es un problema clsico de satisfaccin e restricciones, ideal para ser resuleto mediante la tcnica de vuelta atrs (backtracking)

```python
def escribe(S):
    n=len(S)
    for x in range(n):
        print("")
        for i in range(n):
            if S[i] == x+1:
                print(" X " , end ="")
            else:
                print("  -  ", end="")          
                
        
```

```python
def es_prometedora(SOLUCION, etapa):
    #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
    for i in range(etapa+1):
        if SOLUCION.count(SOLUCION[i])>1:
            return False

        #Verifica las diagonales
        for j in range (i+1, etapa +1):
            if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : 
                return False
    return True
```

```python
def reinas(N, solucion=[], etapa=0):
    if len(solucion)==0:
        solucion =[0 for i in range (N)]
    for i in range(1, N+1):
        solucion[etapa]=i

        if es_prometedora(solucion, etapa):
            if etapa == N-1:
                print(solucion)
                print()
            else:
                reinas(N, solucion, etapa+1)
        else:
            None
        solucion[etapa]=0
    
```
