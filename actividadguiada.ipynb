# Torres de hanoi
## Algoritos recursivos
Un algoritmo recursivo es un mtodo de programacin en que la funcin se invoca a si isma para resolver una versin reducida del mismo problema.
Este proceso contina hasta que el problema se simplifica  lo suficinete como para resolverse directamente, sin necesidad de invocaciones adicionales.
La idea clave de la recursin es que el problema se divide en partes ms pequeas y manejables, lo que hace ms facil resolverlo.

### Ejercicio Torres de Hanoi
Se llama las **Torres de hanoi**. Te dan un conjunto de tres varillas y n discos, con cada disco de un tamao diferente. LLamaremos a las varillas A, B, y numeramos los discos desde 1, siendo el disco ms pequeo, hasta n, siendo el disco ms grande. Al principio, todos los n discos estn en la varilla A, en orden de tamao decreciente de la parte inferior a la parte superior, de modo que el disco n est en la parte inferior  y el disco 1 est en la parte superior. Aqu esta como se ven las torres de Hanoi para n=4 discos:



El objetivo del ejercicio es pasar todos los n discos de la varilla A a la varilla B.


**Debemos seguir las siguientes reglas:**
* Podemos mover solamanete un disco a la vez
* Ningn disco puede estar encima de un disco ms pequeo. Por ejemplo, si el disco 3 est en una varilla, entonces todos los discos debajo del disco 3 deben tener nmeros mayores que 3.

## Pasos a seguir

Siempre puedes mover el disco1 de la varilla A a la varilla B, porqu sabes que cualquier disco debajo debe ser mayor.  Y no hay nada especial acerca de las varillas A y B. ## Trmino general
La sucesin de Fibonacci es una sucesin definida por recurrencia. Esto significa que para calcular un trmino de sucesin se necesitan los trminos que le preceden.<br>
Se proporcionan los dos primeros trminos: $a_0 = 0$ y $a_1 = 1$. Los siguinetes se calculan con la siguinete frmula:<br>

$$
a_n+1 = a_n-1 + a_n, n>= 1
$$
<br>
**Nota:** el primer trmino que proporciona la frmula es **$a_2$** (porque **$n$** tiene que ser mayor o igual que **1**).<br>
Por esta razn, se definen **$a_0$** y **$a_1$** con anterioridad.



```python
#Calculo del trmino n-simo de la sucesin de Fibonacci
def Fibonacci(N:int):
    if N<2:
        return 1
    else:
        return Fibonacci(N-1)+Fibonacci(N-2)
        
```

```python
Fibonacci(5)
```

```python
# Funcin para generar la serie de Fibonacci hasta el trmino n
def serie_fibonacci(N):
    serie = []
    for i in range (N + 1):
        serie.append(Fibonacci(i))
    return serie
```

```python
# Implementamos una funcin para imprimir la serie de fibonacci en seriepiramide
def imprimir_serie_fibonacci(N):
    serie = serie_fibonacci(N)
    for i in range(1, len(serie) +1):
        print (" " * (len(serie)-i), *serie[:i])
```

# Frmula de Binet
**La frmula de Binet** es una frula explicita y cerrada que se utiliza para hallar rl $n$ trmino n-simo de la sucesin de Fibonacci. Recibe este nombre porque fue derivada por el matemtico Jacques Philippe Marie Binet, aunque ya era conocidapor Abraham de Moivre.

**Frmula**
Si $F_n$ es el $n^0$ nmero de Fibonacci, entonces:<br>
$$
F_n = \frac {1}{\sqrt{a}}((\frac{1+\sqrt{5}}{2})^n -(\frac{1-\sqrt{5}}{2})^n)
$$

```python
#importamos la libreria math
import math
def formula_binet(n):
    #calculamos el nmero ureo
    phi = (1 + math.sqrt(5)) / 2
    psi = (1 - math.sqrt(5)) / 2

    resultado = (phi**n - psi**n) / math.sqrt(5)
    return round(resultado)
```

```python
#implementamos una funcin que imprime los primeros n niveles en forma de piramide
def piramide(numero_niveles):
    #Calculamos lso trminos de Fibonnaci necesario
    total_numeros = numero_niveles * (numero_niveles +1)//2
    fibos=[formula_binet(i) for i in range(total_numeros)]
    index = 0
    for fila in range(1, numero_niveles+1):
        #espacios
        print (" " * (numero_niveles-fila) * 3, end="")
        for _ in range(fila):
            print(f"{fibos[index]:<4}", end=" ")
            index+=1
        print()
```
